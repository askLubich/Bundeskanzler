```python
def date2year(date):
    if date == '-':
        now = datetime.datetime.now()
        date = now.strftime("%d.%m.%Y")        
    year = float(date[6:])
    year = year + (float(date[3:5]) -1)/12.
    year = year + (float(date[0:2]) -1)/12./30.    
    return year

def drawBar(ax, i, start, end, args={}):
    ax.barh(i, end-start, left=start, **args)
```

# Plotting

```python
f, ax = plt.subplots(figsize=(18,20))

President2Ndx = {}
i = 0
for president in President:
    name = president.name
    
    if name not in President2Ndx.keys():
        President2Ndx[president.name] = i
        drawBar(ax, President2Ndx[name], date2year(president.born), date2year(president.died), args = {'color': '#ccb3ff', 'alpha': 1.0, 'height': .8})
        ax.text( date2year(president.born)+0.5, i, president.name, ha='left', va='center')
        i = i + 1
    
    if president.party == 'CDU':
        iOcolor = 'black'
    elif president.party == 'SPD':
        iOcolor = 'red'
    elif president.party == 'FDP':
        iOcolor = 'yellow'
    elif president.party == 'NSDAP':
        iOcolor = '#4d1919'
    elif president.party == 'Zentrum':
        iOcolor = 'blue'
    else:
        iOcolor = 'gray'
    
    drawBar(ax, President2Ndx[name], date2year(president.iOstart), date2year(president.iOEnd), args = {'color': iOcolor, 'alpha': 1.0, 'height': .8})

    
Kanzler2Ndx = {}
i = i + 5
for kanzler in Kanzler:
    name = kanzler.name
    
    if name not in Kanzler2Ndx.keys():
        Kanzler2Ndx[kanzler.name] = i
        drawBar(ax, Kanzler2Ndx[name], date2year(kanzler.born), date2year(kanzler.died), args = {'color': '#99ccff', 'alpha': 1.0, 'height': .8})
        ax.text( date2year(kanzler.born)+0.5, i, kanzler.name, ha='left', va='center')
        i = i + 1
    
    if kanzler.party == 'CDU':
        iOcolor = 'black'
    elif kanzler.party == 'SPD':
        iOcolor = 'red'
    elif kanzler.party == 'FDP':
        iOcolor = 'yellow'
    elif kanzler.party == 'NSDAP':
        iOcolor = '#4d1919'
    elif kanzler.party == 'Zentrum':
        iOcolor = 'blue'
    else:
        iOcolor = 'gray'
    
    drawBar(ax, Kanzler2Ndx[name], date2year(kanzler.iOstart), date2year(kanzler.iOEnd), args = {'color': iOcolor, 'alpha': 1.0, 'height': .8})

    
#plot eras
for era in Eras:
    drawBar(ax, -3, date2year(era.start), date2year(era.end), args = {'color': era.color, 'alpha': 1.0, 'height': 2, 'zorder': 1, 'align': 'center'})
    ax.text( date2year(era.start)+1, -3, era.name, ha='left', va='center', color='white')
    
    drawBar(ax, len(President2Ndx)+2, date2year(era.start), date2year(era.end), args = {'color': era.color, 'alpha': 1.0, 'height': 2, 'zorder': 1, 'align': 'center'})
    ax.text( date2year(era.start)+1, len(President2Ndx)+2, era.name, ha='left', va='center', color='white', weight='normal')
    
    drawBar(ax, i+2, date2year(era.start), date2year(era.end), args = {'color': era.color, 'alpha': 1.0, 'height': 2, 'zorder': 1, 'align': 'center'})
    ax.text( date2year(era.start)+1, i+2, era.name, ha='left', va='center', color='white', weight='normal')
    
    drawBar(ax, -4, date2year(era.start), date2year(era.end), args = {'color': era.color, 'alpha': .1, 'height': i+5, 'zorder': -1, 'align': 'edge'})

#labels and legend
drawBar(ax, i+2, 1800, 1850, args = {'color': 'gray', 'alpha': .8, 'height': 2, 'align': 'center'})
ax.text( 1825, i+2, 'Heads of Government', ha='center', va='center', color='white', weight='normal', fontsize=20)

drawBar(ax, len(President2Ndx)+2, 1800, 1850, args = {'color': 'gray', 'alpha': .8, 'height': 2, 'align': 'center'})
ax.text( 1825, len(President2Ndx)+2, 'Heads of State', ha='center', va='center', color='white', weight='normal', fontsize=20)

legendNdx = 32
legendYea = 1965
legendHeight = 8
drawBar(ax, legendNdx-legendHeight+1, legendYea, legendYea+50, args = {'color': 'gray', 'alpha': .8, 'height': legendHeight, 'align': 'edge'})
ax.text(legendYea+2, legendNdx, 'Time in office:', ha='left', va='center', color='white', weight='normal')
ax.text(legendYea+15, legendNdx-1, 'Christian Democratic Union', ha='left', va='center', color='white', weight='normal')
drawBar(ax, legendNdx-1,legendYea+5,legendYea+13, args = {'color': 'black', 'alpha': 1, 'height': .8, 'align': 'center'})
ax.text(legendYea+15, legendNdx-2, 'Social Democratic Party', ha='left', va='center', color='white', weight='normal')
drawBar(ax, legendNdx-2,legendYea+5,legendYea+13, args = {'color': 'red', 'alpha': 1, 'height': .8, 'align': 'center'})
ax.text(legendYea+15, legendNdx-3, 'Free Democratic Party', ha='left', va='center', color='white', weight='normal')
drawBar(ax, legendNdx-3,legendYea+5,legendYea+13, args = {'color': 'yellow', 'alpha': 1, 'height': .8, 'align': 'center'})
ax.text(legendYea+15, legendNdx-4, 'Nazi Party', ha='left', va='center', color='white', weight='normal')
drawBar(ax, legendNdx-4,legendYea+5,legendYea+13, args = {'color': '#4d1919', 'alpha': 1, 'height': .8, 'align': 'center'})
ax.text(legendYea+15, legendNdx-5, 'Center Party', ha='left', va='center', color='white', weight='normal')
drawBar(ax, legendNdx-5,legendYea+5,legendYea+13, args = {'color': 'blue', 'alpha': 1, 'height': .8, 'align': 'center'})
ax.text(legendYea+15, legendNdx-6, 'other/non-partisan', ha='left', va='center', color='white', weight='normal')
drawBar(ax, legendNdx-6,legendYea+5,legendYea+13, args = {'color': 'gray', 'alpha': 1, 'height': .8, 'align': 'center'})
    
#limits, ticks and other stuff
ax.set_xlim(1797, 2018)
ax.set_ylim(-5, i+4)
ax.tick_params(axis='both', left='off', top='on', right='off', bottom='on', labelleft='off', labeltop='on', labelright='off', labelbottom='on')

ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)

#save
plt.savefig('graph.png', bbox_inches='tight', dpi=600)
```
